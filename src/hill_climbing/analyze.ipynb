```python
import json


from hill_climbing.hill_climber import HillClimbingResults
def load_json(filepath) -> list[HillClimbingResults]:
    with open(filepath, 'r') as f:
        data = json.load(f)
    return [HillClimbingResults.model_validate(x) for x in data]
killer = load_json(r"C:\Users\samuel.soukup\Documents\School\AMPEvolve\outputs\hill_climb\evograd_killer.json")
macrel = load_json(r"C:\Users\samuel.soukup\Documents\School\AMPEvolve\outputs\hill_climb\evograd_macrel.json")
evograd = load_json(r"C:\Users\samuel.soukup\Documents\School\AMPEvolve\outputs\hill_climb\evograd.json")
```

```python
killer_end_scores = [x[-1].score for x in killer]
macrel_end_scores = [x[-1].score for x in macrel]
evograd_end_scores = [x[-1].score for x in evograd]
```

```python
import numpy as np
from scipy.stats import ttest_ind

print("K", np.mean(killer_end_scores))
print("M", np.mean(macrel_end_scores))
print("E",np.mean(evograd_end_scores))
# ttest between each other:
for group1, name1 in zip((killer_end_scores, macrel_end_scores, evograd_end_scores), "KME"):
    for group2, name2 in zip((killer_end_scores, macrel_end_scores, evograd_end_scores), "KME"):
        print(f"{name1} X {name2} {ttest_ind(group1, group2)}")
```
